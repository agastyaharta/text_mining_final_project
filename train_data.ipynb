```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data_full = pd.read_csv("drug_review_train.csv", sep=',')
data_full.head(3)
```

```python
take_var_data = data_full[['drugName','condition','review','rating','date','usefulCount']]
take_var_data.info()
```

```python
train_df, temp_df = train_test_split(
    take_var_data,
    test_size=0.40,
    random_state=1,
    shuffle=True
)

# 2. Validation (20%) + Test (20%)
val_df, test_df = train_test_split(
    temp_df,
    test_size=0.50,  
    random_state=1,
    shuffle=True
)
```

```python
train_df.to_csv("train_data.csv", index=False)
val_df.to_csv("val_data.csv", index=False)
test_df.to_csv("test_data.csv", index=False)
```
